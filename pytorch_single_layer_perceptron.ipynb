```python
import torch
```

```python
x=torch.tensor([1.0,2.0])
target=torch.tensor(1.0)
```

```python
weights=torch.tensor([0.2,-0.5],requires_grad=False)
bias=torch.tensor(0.1)
learning_rate=0.1
```

```python
def step_activation(z):
    # Step Activation: 0 if z <= 0, else 1 [00:39:44]
    return 1.0 if z > 0 else 0.0

# 2. TRAINING LOOP (Perceptron Learning Algorithm) [00:31:16]
for epoch in range(3): # The video takes 3 iterations to reach zero error
    # Step A: Weighted Sum (Summation) [00:38:45]
    # z = (x1*w1 + x2*w2) + b
    z = torch.dot(x, weights) + bias

    # Step B: Activation Function [00:39:38]
    y_hat = step_activation(z)

    # Step C: Error Calculation [00:40:49]
    # Error = Target - Computed
    error = target - y_hat

    print(f"Epoch {epoch+1}: z={z:.2f}, Prediction={y_hat}, Error={error}")

    if error == 0:
        print("Converged! Optimal weights found.")
        break

    # Step D: Weight & Bias Updation [00:35:53]
    # W_new = W_old + (LR * error * x)
    weights = weights + (learning_rate * error * x)

    # Bias_new = Bias_old + (LR * error)
    bias = bias + (learning_rate * error)

print(f"\nFinal Weights: {weights}")
print(f"Final Bias: {bias}")
```
